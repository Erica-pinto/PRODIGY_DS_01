```python
################TASK01

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset from the provided link
world_population = pd.read_csv('C:/Users/erica/Desktop/T1.csv', skiprows=4)

# Filter data for a specific country (e.g., United States)
country = 'United States'
population_data_country = world_population.loc[world_population['Country Name'] == country]

# Create a bar chart to visualize population over the years
population_data_country.plot(kind='bar', x='Country Name', figsize=(12, 5), legend=None)

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population Trend for {country}')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the provided link
world_population = pd.read_csv('C:/Users/erica/Desktop/T1.csv', skiprows=4)

# Filter data for a specific country (e.g., United States)
country = 'United States'
population_data_country = world_population.loc[world_population['Country Name'] == country]

# Extract population values
population_values = population_data_country.iloc[:, 4:].values.flatten()

# Create a histogram to visualize population distribution
plt.hist(population_values, bins=20, color='salmon', edgecolor='black')

# Adding labels and title
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title(f'Population Distribution for {country}')

# Display the plot
plt.tight_layout()
plt.show()
```
